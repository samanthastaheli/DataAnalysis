```python
import pandas as pd
import altair as alt

data = pd.read_csv("datasets/friends_episodes.csv")
data.head()
```

```python
imdb_data = pd.read_csv("datasets/friends_imdb.csv")
imdb_data.head()
```

```python
# writers
people = data["written_by"].to_list()

writers_list = []
split_and = []
split_and2 = []
split_s = []
split_st = []
split_and3 = []

# split data
for person in people:
    if "&" in person:
        split_and.append(person.split("&"))
    elif "Story by:" in person:
        split_s.append(person.split("Story by:"))
    else:
        writers_list.append(person)

for p in split_s:
    split_st.append(p[1].split("Teleplay by:"))

for people in split_and:
    for p in people:
        split_and2.append(p.split("Story by:"))

for people in split_and2:
    for p in people:
        split_and3.append(p.split("Teleplay by:"))

# add new values to list
for people in split_and3:
    for p in people:
        writers_list.append(p)
            
for people in split_st:
    for p in people:
        writers_list.append(p)

# remove extra spaces
for person in writers_list:
    person.strip()
```

```python
# writer list to dataset
writers = pd.DataFrame(writers_list, columns=["writer_name"])
```

```python
writers_chart = (alt.Chart(writers).mark_bar().encode(
    alt.X("count(writer_name)", axis={"title":"Episode Amount"}),
    alt.Y("writer_name", sort="-x", axis={"title":"Writer"})
).properties(
    width=500,
    title={"text": "Amount of Episodes Written By"}
))

writer_text = writers_chart.mark_text(
    align='left',
    baseline='middle',
    dx=3
).encode(
    text='count(writer_name):Q'
)

writers_chart + writer_text
```

```python
# episode rating
top10 = imdb_data.sort_values("imdb_rating", ascending=0).iloc[:10]
top5 = imdb_data.sort_values("imdb_rating", ascending=0).iloc[:5]
top3 = imdb_data.sort_values("imdb_rating", ascending=0).iloc[:3]

top10_chart = (alt.Chart(top10).mark_bar(height=15, color="#0DA4C5", opacity=0.75).encode(
    alt.X("imdb_rating"),
    alt.Y("title", sort="-x")
).properties(
    height=250,
    title={"text":"Top 10 Friends Episodes"}
))

top10_text = top10_chart.mark_text(
    align='left',
    baseline='middle',
    dx=3
).encode(
    text='imdb_rating:Q'
)

top10_chart + top10_text
```
